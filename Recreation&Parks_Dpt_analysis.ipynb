```python
airport_df_final.columns
airport_df_final['% Over Base Pay'] = airport_df_final['% Over Base Pay'].astype(float)
airport_df_final.info()
```

```python
airport_mean = airport_df_final.groupby(['Job Class Title', 'Year'])
airport_mean_df= airport_mean.mean()
airport_mean_df = airport_mean_df[['Total Payments', 'Base Pay', 'Average Benefit Cost', 'Payments Over Base Pay', '% Over Base Pay' ]]
airport_mean_df
```

```python
airport_mean_df['% increase Tot Pay'] = airport_mean_df['Total Payments'].groupby(level=0).pct_change()*100
airport_mean_df['% increase Base Pay'] = airport_mean_df['Base Pay'].groupby(level=0).pct_change()*100
airport_mean_df['% increase Benefits'] = airport_mean_df['Average Benefit Cost'].groupby(level=0).pct_change()*100
airport_mean_df['Cum % increase Tot Pay'] =airport_mean_df['% increase Tot Pay'].groupby(level=0).cumsum()
airport_mean_df['Cum % increase Base Pay'] = airport_mean_df['% increase Base Pay'].groupby(level=0).cumsum()
airport_mean_df['Cum % increase Benefits'] = airport_mean_df['% increase Benefits'].groupby(level=0).cumsum()
airport_mean_df
```

```python
airport_mean_df= airport_mean_df.reset_index()
airport_mean_df
```

```python
years = [2014, 2015]
water_data_file = airport_mean_df[airport_mean_df['Year'].isin(years)]
#water_data_file  = water_data_file [['Job Class Title', 'Year', 'Cum % increase Tot Pay', 'Cum % increase Base Pay','Cum % increase Benefits']]
water_data_file 
water_data_file.to_csv('Output/Recreation&parks.csv')
```
