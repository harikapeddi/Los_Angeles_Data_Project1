```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Output/filtered_dataset.csv', low_memory=False)
df
```

```python
grouped_df = df.groupby('Department Title')
grouped_df.nunique()
```

```python
Water_Power_df = df[df['Department Title']=='Water And Power (DWP)']
Water_Power_df.head()
```

```python
Water_Power_df['Year'].unique()
```

```python
Water_Power_df = Water_Power_df[Water_Power_df['Year']!=2016]
Water_Power_df
```

```python
Water_Power_df.isna().sum()
```

```python
year = Water_Power_df.groupby('Year')
year.mean()
```

```python
year = Water_Power_df.groupby('Year')['Projected Annual Salary'].max()
year=pd.DataFrame(year)
year
```

```python
for index, row in year.iterrows()
```

```python
sns.barplot(x='Year',y='Total Payments', data = Water_Power_df)
ax2 = plt.twinx()
sns.barplot(x='Year',y='Projected Annual Salary', data = Water_Power_df, ax=ax2)
```

```python
ax = Water_Power_df.boxplot(x="Year", y="Total Payments", legend=False)
ax2 = ax.twinx()
Water_Power_df.boxplot(x="Year", y="Projected Annual Salary", ax=ax2, legend=False, color="r")
ax.figure.legend()
plt.show()
```

```python
sns.boxplot(x='Year',y='Projected Annual Salary', data = Water_Power_df)
sns.boxplot(x='Year',y='Total Payments', data = Water_Power_df)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
Water_Power_df.boxplot('Total Payments','Year', ax=ax[0])
Water_Power_df.boxplot('Projected Annual Salary','Year', ax=ax[1])
```

```python
dd=pd.melt(Water_Power_df,id_vars=['Year'],value_vars=['Total Payments','Projected Annual Salary'],var_name='test')
sns.boxplot(x='Year',y='value',data=dd,hue='test')
plt.legend(loc='upper left')
print(dd)
```

```python
dd=pd.melt(Water_Power_df,id_vars=['Year'],value_vars=['Total Payments','Base Pay'],var_name='test')
sns.boxplot(x='Year',y='value',data=dd,hue='test')
plt.legend(loc='upper left')
```
