```python
#plotting the job class title 
fig, ax = plt.subplots(figsize=(18,5))
sns.set_palette(sns.color_palette("Paired"))
sns.countplot(x='Job Class Title', data=airport_df_final, order=airport_df_final['Job Class Title'].value_counts().index, hue='Year')
ax.tick_params(rotation=90)
```

```python
#plotting Base pay, Total payments, and Avg Benefots costs by the job class title by year 
fig, ax = plt.subplots(figsize=(15,7))
dd=pd.melt(airport_df_final,id_vars=['Job Class Title'],value_vars=['Total Payments','Base Pay', 'Average Benefit Cost'],var_name='Year')
sns.set_palette(sns.color_palette("Paired"))
sns.barplot(x='Job Class Title',y='value',data=dd,hue='Year')
ax.tick_params(rotation=90)
plt.legend(loc='upper left')
```

```python

fig, ax = plt.subplots(figsize=(15,6))
dd=pd.melt(airport_df_final,id_vars=['Year'],value_vars=['Total Payments','Base Pay', 'Average Benefit Cost'],var_name='Job Class Title')
sns.set_palette(sns.color_palette("Paired"))
sns.barplot(x='Year',y='value',data=dd,hue='Job Class Title')
ax.tick_params(rotation=90)

plt.legend(loc='upper left')
```
